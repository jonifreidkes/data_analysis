# Visualizing The Gender Gap In College Degrees
## Introduction
In this project we will visualize the Gender Gap in College Degrees in USA between 1968 and 2011 in most important majors, including STEM, arts, and library. Several _charjunk_ and _Data-ink Ratio_ considerations will be achieved while making the plots easier to read and analyse.

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
fig = plt.figure(figsize=(9, 18))

for sp in range(0,18,3):
    ind=int(sp/3)
    ax = fig.add_subplot(6,3,(sp+1))
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[ind]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[ind]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.set_title(stem_cats[ind])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if ind == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
        ax.tick_params(labelbottom='off')
    elif ind == 5:
        ax.text(2005, 88, 'Men')
        ax.text(2001, 7, 'Women')
        ax.tick_params(labelbottom='on')
    else:
        ax.tick_params(labelbottom='off')

for sp in range(1,16,3):
    ind=int(sp/3)
    ax = fig.add_subplot(6,3,(sp+1))
    ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[ind]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[ind]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.set_title(lib_arts_cats[ind])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if ind == 0:
        ax.text(2005, 80, 'Men')
        ax.text(2002, 20, 'Women')
        ax.tick_params(labelbottom='off')
    elif ind == 4:
        ax.text(2005, 60, 'Men')
        ax.text(2001, 38, 'Women')
        ax.tick_params(labelbottom='on')
    else:
        ax.tick_params(labelbottom='off')

for sp in range(2,20,3):
    ind=int(sp/3)
    ax = fig.add_subplot(6,3,(sp+1))
    ax.plot(women_degrees['Year'], women_degrees[other_cats[ind]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[ind]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    ax.set_ylim(0,100)
    ax.set_yticks([0,100])
    ax.set_title(other_cats[ind])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if ind == 0:
        ax.text(2005, 90, 'Men')
        ax.text(2002, 6, 'Women')
        ax.tick_params(labelbottom='off')
    elif ind == 5:
        ax.text(2005, 65, 'Men')
        ax.text(2001, 32, 'Women')
        ax.tick_params(labelbottom='on')
    else:
        ax.tick_params(labelbottom='off')
plt.savefig("gender_degrees.png")
plt.show()
```
